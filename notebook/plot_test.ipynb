```python
import tushare as ts
```

```python
api = ts.get_apis()
```

```python
df = ts.bar('300141',conn=api, start_date='2017-10-01')
```

```python
df.head()
```

```python
%pylab inline
df['close'].plot(kind='line',figsize=(20,10))
```

```python
df['ma5']=df['close'].rolling(window=5).mean()
```

```python
df.head(20)
```

```python
df[['close','ma5']].plot(kind='line',figsize=(20,15))
```

```python
df['ma21']=df['close'].rolling(window = 21).mean()
```

```python
df[['close','ma5','ma21']].plot(kind='line',figsize=(20,15))
```
